```python
import requests
```

GET Request

```python
url = 'https://www.usfca.edu/arts-sciences/programs/graduate/data-science'

r = requests.get(url)
r.headers
```

```python
url = "https://www.google.com/search"
payload = {"q":"USF Data Science"}

r = requests.get(url, params=payload)
r.content
```

POST Request

```python
url = 'https://www.w3schools.com/action_page.php'
data = {'key': 'value'}

r = requests.post(url, data=data, headers='')
r.headers
```
