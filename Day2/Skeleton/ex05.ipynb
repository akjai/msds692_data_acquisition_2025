Example 5.
Load from a .docx located at `url`, read text paragraphs and tables

```python
from io import BytesIO
import requests

url = 'https://www.nodc.noaa.gov/archive/arc0090/0150631/1.1/data/0-data/DWH%20SEAMAP%20Groundfish%20Dataset%20Documentation.docx'
```

```python
response = requests.get(url)
data = BytesIO(response.content)
document = Document(data)
# For text
text_output = []
for paragraph in document.paragraphs:
    text_output.append(paragraph.text)

text_output
```

```python
tables = []
for table in document.tables:
    output = []
for row in table.rows:
    row_content = [cell.text for cell in row.cells] # Collect cell content in a list
    output.append(row_content)
    tables += output
tables
```
